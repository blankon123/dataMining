```python
def format_jam23(tanggal, tanggal2):
    #resample yang 23 Jam
    #(total/92*96 - total)/4
    #jadiin fungsi parameter tanggal

    t1 = pd.to_datetime(tanggal)
    t2 = pd.to_datetime(tanggal2)

    cek23Jam = data[(t1 <= data.index) & (t2 > data.index)].ix[:,3:370]
    cek23JamTot = cek23Jam.resample('D',how=['sum'])
    cek23Jam = cek23JamTot/92*96
    cek23Jam = cek23Jam-cek23JamTot
    cek23Jam = cek23Jam/4

    ganti23jam = cek23Jam
    ganti23jam.columns = ganti23jam.columns.droplevel(1)
    ganti23jam = ganti23jam.ix[0]

    t1 = pd.to_datetime(tanggal + ' 01:00:00')
    t2 = pd.to_datetime(tanggal + ' 01:15:00')
    t3 = pd.to_datetime(tanggal + ' 01:30:00')
    t4 = pd.to_datetime(tanggal + ' 01:45:00')
    
    data.loc[t1,'MT_004':'MT_370'] = ganti23jam
    data.loc[t2,'MT_004':'MT_370'] = ganti23jam
    data.loc[t3,'MT_004':'MT_370'] = ganti23jam
    data.loc[t4,'MT_004':'MT_370'] = ganti23jam

def format_jam25(tanggal, tanggal2):
    #resample yang 23 Jam
    #(total - total/100*96)/4
    #jadiin fungsi parameter tanggal

    t1 = pd.to_datetime(tanggal)
    t2 = pd.to_datetime(tanggal2)

    cek25Jam = data[(t1 <= data.index) & (t2 > data.index)].ix[:,3:370]
    cek25JamTot = cek25Jam.resample('D',how=['sum'])
    cek25Jam = cek25JamTot/100*96
    cek25Jam = cek25JamTot-cek25Jam
    cek25Jam = cek25Jam/4

    ganti25jam = cek25Jam
    ganti25jam.columns = ganti25jam.columns.droplevel(1)
    ganti25jam = ganti25jam.ix[0]

    t1 = pd.to_datetime(tanggal + ' 01:00:00')
    t2 = pd.to_datetime(tanggal + ' 01:15:00')
    t3 = pd.to_datetime(tanggal + ' 01:30:00')
    t4 = pd.to_datetime(tanggal + ' 01:45:00')
    
    data.loc[t1,'MT_004':'MT_370'] = ganti25jam
    data.loc[t2,'MT_004':'MT_370'] = ganti25jam
    data.loc[t3,'MT_004':'MT_370'] = ganti25jam
    data.loc[t4,'MT_004':'MT_370'] = ganti25jam
```

```python
#23jam(27,25,31,30) Maret
format_jam23('2011-03-27', '2011-03-28')
format_jam23('2012-03-25', '2012-03-26')
format_jam23('2013-03-31', '2013-04-01')
format_jam23('2014-03-30', '2014-03-31')

#25jam(30,28,27,26) Oktober
format_jam25('2011-10-30', '2011-11-01')
format_jam25('2012-10-28', '2012-10-29')
format_jam25('2013-10-27', '2013-10-28')
format_jam25('2014-10-26', '2014-10-27')
```

```python
start_time = time.time()
langganan = pd.Series()

for i in data.columns:
    for j in range(0,len(data[i])-1):
        if(data[i][j] > 0):
            langganan[i]=data[i].index[j]
            break
        else:
            continue
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
#resample pelanggan
start_time = time.time()
mthWeekendPagi = pd.DataFrame()
mthWeekendMlm  = pd.DataFrame()
mthWeekdayPagi = pd.DataFrame()
mthWeekdayMlm  = pd.DataFrame()

dataKhusus = pd.DataFrame(columns=['std','qrt1','qrt3','itQ','bukanOutlier'],index=data.columns)
dataRata = pd.DataFrame(columns=['Weekend-Pagi','Weekend-Malam','Weekday-pagi','Weekday-Malam'],index=data.columns)
dataSkew = pd.DataFrame(columns=['Weekend-Pagi','Weekend-Malam','Weekday-pagi','Weekday-Malam'],index=data.columns)
dataKurt = pd.DataFrame(columns=['Weekend-Pagi','Weekend-Malam','Weekday-pagi','Weekday-Malam'],index=data.columns)
dataStd = pd.DataFrame(columns=['Weekend-Pagi','Weekend-Malam','Weekday-pagi','Weekday-Malam'],index=data.columns)

for i in data.columns:
    mth  = pd.DataFrame()
    mth  = data[i][data.index >= langganan[i]]
    mth  = mth.resample('H',how = ['sum'])
    
    mthWeekendPagi = mth['sum'][(mth.index.weekday >= 5) & (mth.index.weekday <= 6)].between_time(datetime.time(7,0),datetime.time(20,0))
    mthWeekendMlm  = mth['sum'][(mth.index.weekday >= 5) & (mth.index.weekday <= 6)].between_time(datetime.time(20,0),datetime.time(7,0))
    mthWeekdayPagi = mth['sum'][(mth.index.weekday >= 0) & (mth.index.weekday <= 4)].between_time(datetime.time(7,0),datetime.time(20,0))
    mthWeekdayMlm  = mth['sum'][(mth.index.weekday >= 0) & (mth.index.weekday <= 4)].between_time(datetime.time(20,0),datetime.time(7,0))

    dataRata['Weekend-Pagi'][i]  = mthWeekendPagi.mean()
    dataRata['Weekend-Malam'][i] = mthWeekendMlm.mean()
    dataRata['Weekday-pagi'][i]  = mthWeekdayPagi.mean()
    dataRata['Weekday-Malam'][i] = mthWeekdayMlm.mean()

    dataSkew['Weekend-Pagi'][i]  = mthWeekendPagi.skew()
    dataSkew['Weekend-Malam'][i] = mthWeekendMlm.skew()
    dataSkew['Weekday-pagi'][i]  = mthWeekdayPagi.skew()
    dataSkew['Weekday-Malam'][i] = mthWeekdayMlm.skew()

    dataKurt['Weekend-Pagi'][i]  = mthWeekendPagi.kurtosis()
    dataKurt['Weekend-Malam'][i] = mthWeekendMlm.kurtosis()
    dataKurt['Weekday-pagi'][i]  = mthWeekdayPagi.kurtosis()
    dataKurt['Weekday-Malam'][i] = mthWeekdayMlm.kurtosis()
    
    dataStd['Weekend-Pagi'][i]  = mthWeekendPagi.var()**(.5)
    dataStd['Weekend-Malam'][i] = mthWeekendMlm.var()**(.5)
    dataStd['Weekday-pagi'][i]  = mthWeekdayPagi.var()**(.5)
    dataStd['Weekday-Malam'][i] = mthWeekdayMlm.var()**(.5)
    
    dataKhusus['mean'][i]  = mth['sum'].mean()
    dataKhusus['std'][i]  = mth['sum'].var()
    dataKhusus['qrt1'][i] = mth['sum'].quantile(0.25)
    dataKhusus['qrt3'][i] = mth['sum'].quantile(0.75)
    dataKhusus['itQ'][i]  = dataKhusus['qrt3'][i] - dataKhusus['qrt1'][i]
    
    btBwah = dataKhusus['qrt1'][i] - (1.5*dataKhusus['itQ'][i])
    btAtas = dataKhusus['qrt3'][i] + (1.5*dataKhusus['itQ'][i])
    totalData = len(mth['sum'])
    bukanOutlier = len(mth['sum'][ (mth['sum']>=btBwah) & (mth['sum']<=btAtas) ])
    
    dataKhusus['bukanOutlier'][i] = float(bukanOutlier)/float(totalData)
    
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
frames = [dataRata, dataSkew, dataKurt, dataStd]
dataClean = pd.concat(frames, keys=['Rata', 'Skew', 'Kurt','Std'])
dataClean = dataClean.unstack(level=0)

dataClean.head()
```

```python
import sklearn
from scipy.spatial.distance import cdist,pdist
from sklearn.cluster import KMeans
```

```python
hpc = dataClean.values
kmeans = KMeans()
kmeans.fit(hpc)
```

```python
#menentukan explained varians , makin besar maik baik
k_range = range(1,10)
k_means_var = [KMeans(n_clusters=k).fit(hpc) for k in k_range]
centro = [X.cluster_centers_ for X in k_means_var]
k_euclid = [cdist(hpc,cent,'euclidean') for cent in centro]
dist = [np.min(ke,axis=1) for ke in k_euclid]
wcss = [sum(d**2) for d in dist]
tss = sum(pdist(hpc)**2)/hpc.shape[0]
bss = tss - wcss

pp = pd.Series(100*bss/tss)
          
ax1 = plt.subplot(1,1,1)
pp.plot()

for i,j in zip(pp.index,pp):
    ax1.annotate(str(j),xy=(i,j))
plt.scatter(pp.index,pp)
plt.grid()
plt.legend()
plt.show()
```

```python
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.cm as cm

X = hpc
y = hpc
range_n_clusters = [2, 3, 4, 5, 6]

for n_cluster in range_n_clusters:
    # Create a subplot with 1 row and 2 columns
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)

    # The 1st subplot is the silhouette plot
    # The silhouette coefficient can range from -1, 1 but in this example all
    # lie within [-0.1, 1]
    ax1.set_xlim([-0.1, 1])
    # The (n_clusters+1)*10 is for inserting blank space between silhouette
    # plots of individual clusters, to demarcate them clearly.
    ax1.set_ylim([0, len(X) + (n_clusters + 1) * 10])

    # Initialize the clusterer with n_clusters value and a random generator
    # seed of 10 for reproducibility.
    clusterer = KMeans(n_clusters=n_cluster)
    cluster_labels = clusterer.fit_predict(X)

    # The silhouette_score gives the average value for all the samples.
    # This gives a perspective into the density and separation of the formed
    # clusters
    silhouette_avg = silhouette_score(X, cluster_labels)
    print("For n_clusters =", n_clusters,
          "The average silhouette_score is :", silhouette_avg)

    # Compute the silhouette scores for each sample
    sample_silhouette_values = silhouette_samples(X, cluster_labels)

    y_lower = 10
    for i in range(n_clusters):
        # Aggregate the silhouette scores for samples belonging to
        # cluster i, and sort them
        ith_cluster_silhouette_values = \
            sample_silhouette_values[cluster_labels == i]

        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.spectral(float(i) / n_clusters)
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)

        # Label the silhouette plots with their cluster numbers at the middle
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Compute the new y_lower for next plot
        y_lower = y_upper + 10  # 10 for the 0 samples

    ax1.set_title("The silhouette plot for the various clusters.")
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster label")

    # The vertical line for average silhoutte score of all the values
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    ax1.set_yticks([])  # Clear the yaxis labels / ticks
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    # 2nd Plot showing the actual clusters formed
    colors = cm.spectral(cluster_labels.astype(float) / n_clusters)
    ax2.scatter(X[:, 0], X[:, 1], marker='.', s=30, lw=0, alpha=0.7,
                c=colors)

    # Labeling the clusters
    centers = clusterer.cluster_centers_
    # Draw white circles at cluster centers
    ax2.scatter(centers[:, 0], centers[:, 1],
                marker='o', c="white", alpha=1, s=200)

    for i, c in enumerate(centers):
        ax2.scatter(c[0], c[1], marker='$%d$' % i, alpha=1, s=50)

    ax2.set_title("The visualization of the clustered data.")
    ax2.set_xlabel("Feature space for the 1st feature")
    ax2.set_ylabel("Feature space for the 2nd feature")

    plt.suptitle(("Silhouette analysis for KMeans clustering on sample data "
                  "with n_clusters = %d" % n_clusters),
                 fontsize=14, fontweight='bold')

    plt.show()
```
