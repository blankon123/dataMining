```python
#resample yang 23 Jam
#(di nolin)(total/92*96 - total)/4
cek23Jam = data[(data.index >= '2012-03-25 00:00:00') & (data.index < '2012-03-26 00:00:00')].ix[:,3:370]
cek23JamTot = cek23Jam.resample('D',how=['sum'])
cek23Jam = cek23JamTot/92*96
cek23Jam = cek23Jam-cek23JamTot
cek23Jam = cek23Jam/4
cek23Jam
```
